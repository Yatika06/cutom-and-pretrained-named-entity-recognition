```python
def preprocess(text):
  txt_new = re.sub('[^A-Za-z0-9]+', ' ', text)
  txt_new = word_tokenize(txt_new)
  txt_new = nltk.pos_tag(txt_new)
  return txt_new


sent = preprocess(txt)
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}' #Our chunk pattern consists of one rule, that a noun phrase, NP, should be formed whenever the chunker finds an optional determiner, DT, followed by any number of adjectives, JJ, and then a noun, NN.
```

```python
cp = nltk.RegexpParser(pattern)
cs = cp.parse(sent)
print(cs)
```

```python
iob_tagged = tree2conlltags(cs)
pprint(iob_tagged)
```

## Script to train and test the model

```python
from __future__ import unicode_literals, print_function
import plac
import random
from pathlib import Path
import spacy
from tqdm import tqdm
```

```python
model = None
n_iter= 50
output_dir = ('/content/NER_out')
```

```python
#Load the model
if model is not None:
  nlp= spacy.load(model)
  print("Loaded model '%s'" % model)
else:
  nlp = spacy.blank('en')  
  print("Created blank 'en' model")
```

```python
#set up the pipeline
if 'ner' not in nlp.pipe_names:
  ner = nlp.create_pipe('ner')
  nlp.add_pipe(ner, last = True)
else:
  ner = nlp.get_pipe('ner')
```

```python
from spacy.util import minibatch, compounding
```

```python
#train the recognizer
for _, annotations in Train_data[:20000]:
    for ent in annotations.get('entities'):
      ner.add_label(ent[2])

if model is None:
    optimizer = nlp.begin_training()
else:
    optimizer = nlp.entity.create_optimizer()

other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
with nlp.disable_pipes(*other_pipes):  # only train NER
    for itn in range(n_iter):
        random.shuffle(Train_data[:20000])
        losses = {}
        batches = minibatch(Train_data[:20000], 
                            size=compounding(4., 32., 1.001))
        for batch in batches:
            texts, annotations = zip(*batch) 
            # Updating the weights
            nlp.update(texts, annotations, sgd=optimizer, 
                       drop=0.35, losses=losses)
        print('Losses', losses)
```

```python
for text, _ in Train_data[:100]:
    doc = nlp(text)
    print('Entities', [(ent.text, ent.label_) for ent in doc.ents])
    print('Tokens', [(t.text, t.ent_type_, t.ent_iob) for t in doc])
```
